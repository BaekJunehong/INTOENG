## 피클파일 모델 실행

```python
import pickle
from transformers import pipeline

# 피클 파일에서 ASR 파이프라인 불러오기
with open('/content/drive/MyDrive/ComprehensiveDesign_modeling/model/whisper_small_0428.pkl', 'rb') as f:
    asr_loaded = pickle.load(f)

# 불러온 ASR 파이프라인을 이용해 음성 파일 처리 예시
# 여기서는 음성 파일 경로 리스트와 transcribe_audio 함수를 사용합니다.

audio_file_paths = [
    "/content/drive/MyDrive/ComprehensiveDesign_modeling/data/audio/100038_audio.wav",
    "/content/drive/MyDrive/ComprehensiveDesign_modeling/data/audio/100114_audio.wav",
    "/content/drive/MyDrive/ComprehensiveDesign_modeling/data/audio/100128_audio.wav",
    # 추가 음성 파일 경로
]

# ASR 함수 정의 (불러온 asr_loaded 파이프라인 사용)
def transcribe_audio(audio_path):
    transcription = asr_loaded(audio_path)
    return transcription['text']  # Use 'text' key to get the transcribed text

# 각 음성 파일에 대한 처리 및 출력
for audio_file_path in audio_file_paths:
    transcription_text = transcribe_audio(audio_file_path)
    print(f"음성 파일: {audio_file_path}")
    print("텍스트 출력:", transcription_text)
    print()
```

### 원본파일 모델 실행

```python
from transformers import pipeline

# ASR 파이프라인 초기화
model_name_or_path = "/content/drive/MyDrive/ComprehensiveDesign_modeling/model/whisper_small_0428"
asr = pipeline(model=model_name_or_path, task="automatic-speech-recognition")

# 음성 파일 경로 리스트
audio_file_paths = [
    "/content/drive/MyDrive/ComprehensiveDesign_modeling/data/audio/100038_audio.wav",
    "/content/drive/MyDrive/ComprehensiveDesign_modeling/data/audio/100114_audio.wav",
    "/content/drive/MyDrive/ComprehensiveDesign_modeling/data/audio/100128_audio.wav",
    # 추가 음성 파일 경로
]

# ASR 함수 정의
def transcribe_audio(audio_path):
    transcription = asr(audio_path)
    return transcription['text']  # Use 'text' key to get the transcribed text

# 각 음성 파일에 대한 처리 및 출력
for audio_file_path in audio_file_paths:
    transcription_text = transcribe_audio(audio_file_path)
    print(f"음성 파일: {audio_file_path}")
    print("텍스트 출력:", transcription_text)
    print()
```
