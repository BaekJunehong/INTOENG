### 전반적인 모델에 대한 시각화

```python
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 준비
steps = [80, 160, 240, 320, 400, 480]

# 각 모델별 데이터
models_data = {
    'tiny': {
        'train_losses': [3.509800, 1.529500, 1.374600, 1.311500, 1.223700, 1.223700],
        'val_losses': [1.542209, 1.417196, 1.382577, 1.367970, 1.362651, 1.362651],
        'wer_list': [0.569914, 0.523725, 0.536093, 0.549218, 0.589601, 0.549470]
    },
    'base': {
        'train_losses': [2.540400, 1.473500, 1.312300, 1.243000, 1.146400, 1.132800],
        'val_losses': [1.494949, 1.386888, 1.358901, 1.348510, 1.344508, 1.342330],
        'wer_list': [0.546189, 0.591620, 0.637809, 0.636547, 0.582029, 0.630994]
    },
    'small': {
        'train_losses': [3.651400, 1.547800, 1.183700, 1.092600, 0.930500, 0.896300],
        'val_losses': [1.895640, 1.313774, 1.284620, 1.276894, 1.275117, 1.274415],
        'wer_list': [0.512115, 0.504038, 0.506815, 0.492680, 0.542403, 0.483342]
    },
    'medium': {
        'train_losses': [3.690800, 1.311900, 1.058200, 0.959300, 0.756900, 0.705200],
        'val_losses': [1.329034, 1.244087, 1.232604, 1.231918, 1.234329, 1.256282],
        'wer_list': [0.518173, 0.499748, 0.519182, 0.531045, 0.536850, 0.476527]
    }
}

# 그래프 생성 (1, 3 배치)
fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(18, 6))

# 첫 번째 그래프 (Train Loss)
for model_name, data in models_data.items():
    ax1.plot(steps, data['train_losses'], label=f'{model_name} Train Loss')
ax1.set_title('Train Loss over Steps')
ax1.set_xlabel('Steps')
ax1.set_ylabel('Loss')
ax1.set_ylim([0, 5])  # y축의 범위를 0에서 5로 설정
ax1.legend()

# 두 번째 그래프 (Validation Loss)
for model_name, data in models_data.items():
    ax2.plot(steps, data['val_losses'], label=f'{model_name} Validation Loss')
ax2.set_title('Validation Loss over Steps')
ax2.set_xlabel('Steps')
ax2.set_ylabel('Loss')
ax2.set_ylim([0, 5])  # y축의 범위를 0에서 5로 설정
ax2.legend()

# 세 번째 그래프 (WER)
for model_name, data in models_data.items():
    ax3.plot(steps, data['wer_list'], label=f'{model_name} WER')
ax3.set_title('WER over Steps')
ax3.set_xlabel('Steps')
ax3.set_ylabel('WER')
ax3.set_ylim([0, 1])  # y축의 범위를 0에서 1로 설정
ax3.legend()

# 전체 그래프에 대한 제목 설정
fig.suptitle('Whisper Model Performance Comparison', fontsize=18)

plt.tight_layout()
plt.show()
```
